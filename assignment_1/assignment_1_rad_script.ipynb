```python
%matplotlib inline
import pysd
import pandas as pd
import numpy as np
```

```python
PySD_model = pysd.read_vensim('models/ASD.mdl')

def RunASDModel(persuasion_rate=0.05, deradicalization_rate=0.02):
    SD_result = PySD_model.run(params={
                            'persuasion rate':persuasion_rate,
                            'deradicalization rate':deradicalization_rate
                                       })
    time = SD_result.index.values
    return SD_result['radicalized citizens']
```

```python
scenarios = map(lambda s: RunASDModel((s/10.0), 0.002), range(1,5))
```

```python
i = 1
for scenario in scenarios:
    scenario.plot(legend=True, label="Scenario " + str(i))
    i += 1
```
